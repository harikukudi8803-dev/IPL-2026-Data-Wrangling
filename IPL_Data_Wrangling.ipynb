```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df['batting_team'] = df['batting_team'].str.strip()
df['bowling_team'] = df['bowling_team'].str.strip()
df['venue'] = df['venue'].str.strip()
df['striker'] = df['striker'].str.strip()
df['bowler'] = df['bowler'].str.strip()
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day_name()
```

```python
df['total_runs'] = df['runs_of_bat'] + df['extras']
```

```python
df.to_csv("IPL_2026_Cleaned.csv", index=False)
```
